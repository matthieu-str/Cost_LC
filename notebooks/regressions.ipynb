```python
# For non-linear regression, we use the Gekko library

from gekko import GEKKO
import numpy as np
import matplotlib.pyplot as plt  
from scipy import stats
```

Gekko documentation: https://gekko.readthedocs.io/en/stable/index.html

Wright's learning curve form : $\log(c) = a + b.\log(Q)$ i.e. $c = \exp(a).Q^b$ 

By setting $A = \exp(a)$, we get $c = A.Q^b$ which is a power law. 

$LR = 1 - 2^b$ is the learning rate. 

# Onshore Wind

## $\log(c) = a + b.\log(Q)$

```python
# data
cost = np.array([1914,1883,1937,1797,1745,1611,1604,1599,1520,1448,1316])
cap = np.array([178,216,262,293,341,404,453,495,540,593,698])

# Gekko model
m = GEKKO()

# parameters
x = m.Param(value=cap)
a = m.FV()
a.STATUS=1
b = m.FV()
b.STATUS=1

# variables
y = m.CV(value=cost)
y.FSTATUS=1

# regression equation
m.Equation(m.log(y)==a+b*m.log(x))

# regression mode
m.options.IMODE = 2

# optimize
m.solve(disp=False)

# print parameters
print(a.value[0], b.value[0])

plt.plot(cap, cost,'bo',label='data')
plt.plot(cap, y.value,'r-',label='regression')
plt.loglog()
plt.xlabel('Cumulative capacity [GW]')
plt.ylabel('Cost [USD_2018/kW]')
plt.legend()
plt.show()

slope, intercept, r_value, p_value, std_err = stats.linregress(cost,y)

r2 = r_value**2 
print("R^2 = " + str(r_value**2))
```

```python
# Testing if the prospective cost stays positive
cap_2050 = 8265*0.8345
cost_2050 = np.exp(a[0]+b[0]*np.log(cap_2050))
print(cost_2050)
```

```python
# Learning rate 
LR = 1 - 2**(b[0])
print(round(LR,4))
```
